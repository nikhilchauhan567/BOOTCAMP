# Problem Description
The objective of this analysis is to gain insights into the characteristics of colleges and answer key questions related to the educational landscape.<br> By understanding the data, we aim to inform strategies for improving the quality of education and enhancing the overall college experience.<br> The analysis will provide valuable insights and recommendations for stakeholders in the education sector.<br>
**Data Description**<br>
•	**Names**: Names of various university and colleges<br>
•	**Apps**: Number of applications received<br>
•	**Accept**: Number of applications accepted<br>
•	**Enroll**: Number of new students enrolled<br>
•	**Top10perc**: Percentage of new students from top 10% of Higher Secondary class<br>
•	**Top25perc**: Percentage of new students from top 25% of Higher Secondary class<br>
•	**F.Undergrad**: Number of full-time undergraduate students<br>
•	**P.Undergrad**: Number of part-time undergraduate students<br>
•	**Outstate**: Number of students for whom the particular college or university is Out-of-state tuition<br>
•	**Room.Board**: Cost of Room and board<br>
•	**Books**: Estimated book costs for a student<br>
•	**Personal**: Estimated personal spending for a student<br>
•	**PhD**: Percentage of faculties with Ph.D.’s<br>
•	**Terminal**: Percentage of faculties with terminal degree<br>
•	**S.F.Ratio**: Student/faculty ratio<br>
•	**perc.alumni**: Percentage of alumni who donate<br>
•	**Expend**: The Instructional expenditure per student<br>
•	**Grad.Rate**: Graduation rate<br>

# Application and Enrollment Analysis

**What is the average number of applications received by colleges?**

```python
avg = round(df['Apps'].mean(),3)
print("average number of applications received by colleges:",avg)
```

**What percentage of applications are accepted on average across all colleges?**

```python
df['Accept_Rate']=(df['Accept']/df['Apps'])*100
avg = round(df['Accept_Rate'].mean(),2)
print("percentage of applications are accepted on average across all colleges:",avg)
```

**What is the average enrollment rate?**

```python
df['Enroll_rate']=(df['Enroll']/df['Accept'])*100
avg =round(df['Enroll_rate'].mean(),2)
print("Average Enrollment Rate:",avg)
```

## Academic Excellence

```python
# Calculate the average percentage of new students from the top 10% of their higher secondary class
avg_top10perc = df['Top10perc'].mean()

# Calculate the average percentage of new students from the top 25% of their higher secondary class
avg_top25perc = df['Top25perc'].mean()

print("Average percentage of new students from the top 10% of their higher secondary class",avg_top10perc)
print("Average percentage of new students from the top 25% of their higher secondary class",avg_top25perc)

# Plotting the scatter plot for Top10perc vs Top25perc
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Top10perc', y='Top25perc', data=df)
plt.title('Correlation between Top 10% and Top 25% of High School Class')
plt.xlabel('Percentage of Students from Top 10%')
plt.ylabel('Percentage of Students from Top 25%')
plt.grid(True)
plt.show()
```

## Student Demographics

```python
# Calculate the average number of full-time undergraduate students per college
avg_full_time_undergrad = df['F.Undergrad'].mean()

# Calculate the average number of part-time undergraduate students per college
avg_part_time_undergrad = df['P.Undergrad'].mean()

# Find the college with the highest number of out-of-state students
college_highest_outstate = df.loc[df['Outstate'].idxmax(), 'Names']

print("Average number of full-time undergraduate students per college:" ,avg_full_time_undergrad)
print("Average number of part-time undergraduate students per college:" ,avg_part_time_undergrad)
print("College with the highest number of out-of-state students: ", college_highest_outstate)
```

## Cost and Spending

```python
# Calculate the average cost of room and board across all colleges
avg_room_board = df['Room.Board'].mean()

# Calculate the average estimated book cost for a student
avg_book_cost = df['Books'].mean()

# Calculate the average estimated personal spending for a student
avg_personal_spending = df['Personal'].mean()

# Calculate the variation of instructional expenditure per student across colleges
instructional_expenditure_variation = df['Expend'].describe()

# Print the results
print(f"Average cost of room and board:{avg_room_board:.2f}")
print(f"Average estimated book cost: {avg_book_cost:.2f}")
print(f"Average estimated personal spending: {avg_personal_spending:.2f}")
print("Variation of instructional expenditure per student:")
print(instructional_expenditure_variation)
```

## Faculty Qualification

```python
# Average percentage of faculties with Ph.D.s across all colleges
avg_phd_faculty = df['PhD'].mean()
print(f"Average percentage of faculties with Ph.D.s: {avg_phd_faculty:.2f}%")

# Average percentage of faculties with terminal degrees across all colleges
avg_terminal_faculty = df['Terminal'].mean()
print(f"Average percentage of faculties with terminal degrees: {avg_terminal_faculty:.2f}%")

# Correlation between the percentage of faculties with Ph.D.s and the graduation rate
correlation_phd_grad_rate = df['PhD'].corr(df['Grad.Rate'])
print(f"Correlation between percentage of faculties with Ph.D.s and graduation rate: {correlation_phd_grad_rate:.2f}")
# Plotting the scatter plot with a regression line for percentage of faculties with Ph.D.s vs graduation rate
plt.figure(figsize=(10, 6))
sns.regplot(x='PhD', y='Grad.Rate', data=df, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.title('Correlation between Percentage of Faculties with Ph.D.s and Graduation Rate')
plt.xlabel('Percentage of Faculties with Ph.D.s (%)')
plt.ylabel('Graduation Rate (%)')
plt.grid(True)
plt.show()
```

## Student-Faculty Interaction

```python
# Average student/faculty ratio across all colleges
avg_student_faculty_ratio = df['S.F.Ratio'].mean()
print(f"Average student/faculty ratio: {avg_student_faculty_ratio:.2f}")

# College with the lowest student/faculty ratio
college_lowest_sfratio = df.loc[df['S.F.Ratio'].idxmin(), 'Names']
print(f"College with the lowest student/faculty ratio: {college_lowest_sfratio}")

# Correlation between the student/faculty ratio and the graduation rate
correlation_sfratio_grad_rate = df['S.F.Ratio'].corr(df['Grad.Rate'])
print(f"Correlation between student/faculty ratio and graduation rate: {correlation_sfratio_grad_rate:.2f}")
# Plotting the scatter plot with a regression line for student/faculty ratio vs graduation rate
plt.figure(figsize=(10, 6))
sns.regplot(x='S.F.Ratio', y='Grad.Rate', data=df, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.title('Correlation between Student/Faculty Ratio and Graduation Rate')
plt.xlabel('Student/Faculty Ratio')
plt.ylabel('Graduation Rate (%)')
plt.grid(True)
plt.show()
```

## Alumni Engagement

```python
# Average percentage of alumni who donate across all colleges
avg_alumni_donate = df['perc.alumni'].mean()
print(f"Average percentage of alumni who donate: {avg_alumni_donate:.2f}%")

# Correlation between the percentage of alumni who donate and the graduation rate
correlation_alumni_grad_rate = df['perc.alumni'].corr(df['Grad.Rate'])
print(f"Correlation between percentage of alumni who donate and graduation rate: {correlation_alumni_grad_rate:.2f}")
# Plotting the scatter plot with a regression line for percentage of alumni who donate vs graduation rate
plt.figure(figsize=(10, 6))
sns.regplot(x='perc.alumni', y='Grad.Rate', data=df, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.title('Correlation between Percentage of Alumni Who Donate and Graduation Rate')
plt.xlabel('Percentage of Alumni Who Donate (%)')
plt.ylabel('Graduation Rate (%)')
plt.grid(True)
plt.show()
```

## Graduation Rates

```python
# Average graduation rate across all colleges
avg_graduation_rate = df['Grad.Rate'].mean()
print(f"Average graduation rate: {avg_graduation_rate:.2f}%")

# College with the highest graduation rate
college_highest_grad_rate = df.loc[df['Grad.Rate'].idxmax(), 'Names']
print(f"College with the highest graduation rate: {college_highest_grad_rate}")

# Correlation between the instructional expenditure per student and the graduation rate
correlation_expend_grad_rate = df['Expend'].corr(df['Grad.Rate'])
print(f"Correlation between instructional expenditure per student and graduation rate: {correlation_expend_grad_rate:.2f}")
# Plotting the scatter plot with a regression line for instructional expenditure per student vs graduation rate
plt.figure(figsize=(10, 6))
sns.regplot(x='Expend', y='Grad.Rate', data=df, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.title('Correlation between Instructional Expenditure per Student and Graduation Rate')
plt.xlabel('Instructional Expenditure per Student ($)')
plt.ylabel('Graduation Rate (%)')
plt.grid(True)
plt.show()
```

## Overall Insights

```python
# Calculate correlations between graduation rate and various factors
correlation_matrix = df[['Apps', 'Accept', 'Enroll', 'Top10perc', 'Top25perc', 'Room.Board', 'Books', 'Personal', 'PhD', 'Terminal', 'S.F.Ratio', 'perc.alumni', 'Expend', 'Grad.Rate']].corr()

# Extract only the correlations with graduation rate
correlation_with_grad_rate = correlation_matrix[['Grad.Rate']].drop('Grad.Rate')

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_with_grad_rate, annot=True, cmap='coolwarm', cbar=True, vmin=-1, vmax=1)
plt.title('Correlation of Various Factors with Graduation Rate')
plt.show()
```

**Recomendations**

Based on the correlations, we can make recommendations to colleges on how to improve their graduation rates. Higher positive correlation values indicate stronger associations with graduation rates.

Enhance Academic Excellence: If correlations with Top10perc and Top25perc are high, colleges should focus on admitting students who performed well in their higher secondary classes.

Increase Faculty Qualifications: If PhD and Terminal have high correlations, hiring more qualified faculty members with Ph.D.s and terminal degrees could improve graduation rates.

Improve Student-to-Faculty Ratio: A high correlation with S.F.Ratio suggests that maintaining a lower student-to-faculty ratio can positively impact graduation rates, indicating more personalized attention for students.

Boost Alumni Engagement: If perc.alumni shows a strong positive correlation, increasing alumni donations and engagement can create better facilities and programs for students, thus improving graduation rates.

Increase Instructional Expenditure: A high correlation with Expend indicates that higher spending on instructional materials and resources per student.
